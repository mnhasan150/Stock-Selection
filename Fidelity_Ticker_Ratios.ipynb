```python
# pip install yfinance --upgrade --no-cache-dir

# import modules
import pandas as pd
import yfinance as yf

# import time to calculate execution runtime
import time
from datetime import timedelta
  
# load excel with its path
# wrkbk = op.load_workbook("all_tickers.xlsx")
  
# sh = wrkbk.active

# List of Excel worksheets for each sector of stocks
tech     = 'technology_tickers.xlsx' # FTEC Technology stocks list
health   = 'healthcare_tickers.xlsx' # FHLC Healthcare stocks list
staples  = 'FSTA_tickers.xlsx'       # FSTA Consumer Staples stocks list
disc     = 'FDIS_tickers.xlsx'       # FDIS Consumer Discretionary stocks list
telecom  = 'FCOM_tickers.xlsx'       # FCOM Telecommunication Services stocks list
energy   = 'energy_tickers.xlsx'     # FENY Energy stocks list
estate   = 'FREL_tickers.xlsx'       # FREL Real Estate stock List
utility  = 'FUTY_tickers.xlsx'       # FUTY Utility stock list
filtered = 'filtered_tickers.xlsx'   # Filtered list of stocks for Debt Ratio
fidelity = 'fidelity_tickers.xlsx'   # All Fidelity listed tickers
market   = 'all_tickers.xlsx'        # All tickers from NYSE, NASDAQ, and AMEX
hedge    = 'hedge_fund_tickers.xlsx' # Most bought by activist hedge funds
penny    = 'penny_stock_tickers.xlsx'# Most active penny stocks

# Change the sector name inside the parenthesis
df = pd.read_excel(tech) # can also index sheet by name or fetch all sheets
list = df['Ticker'].tolist()
```

```python
# Start calculating runtime
start_time = time.monotonic()


print('Ticker', 'MC', 'AV', 'DE', 'BETA', 'PM', 'ROA', 'PR', 'DY', 'CR', 'RM', 'TPE', 'FPE', 'PEG', 'PB', 'TMP', 'YL', 'YH', 'CP') # See below:
for item in list:
    switch = yf.Ticker(item)
    try:
        MC = switch.info['marketCap'] # Market Capital
    except:
        pass
    try:
        AV = switch.info['averageVolume'] # Average Volume
    except:
        pass
    try:
        DE = switch.info['debtToEquity'] # Debt-Equity Ratio
    except:
        pass
    try:
        BETA = switch.info['beta'] # Beta
    except:
        pass
    try:
        PM = switch.info['profitMargins'] # Profit Margin
    except:
        pass    
    try:
        ROA = switch.info['returnOnAssets'] # Return on Assets
    except:
        pass
    try:
        PR = switch.info['payoutRatio'] # Payout Ratio
    except:
        pass
    try:
        DY = switch.info['dividendYield'] # Dividend Yield
    except:
        pass
    try:
        CR = switch.info['currentRatio'] # Current Ratio
    except:
        pass
    try:
        RM = switch.info['recommendationMean'] # Recommendation Mean Price
    except:
        pass  
    try:
        TPE = switch.info['trailingPE'] # Trailing PE Ratio
    except:
        pass      
    try:
        FPE = switch.info['forwardPE'] # Forward PE Ratio
    except:
        pass     
    try:
        PEG = switch.info['pegRatio'] # PEG Ratio
    except:
        pass    
    try:
        PB = switch.info['priceToBook'] # Price to Book Ratio
    except:
        pass       
    try:
        TMP = switch.info['targetMeanPrice'] # Analyst Target Mean Price
    except:
        pass
    try:
        YL = switch.info['fiftyTwoWeekLow'] # 52 Week Low Price
    except:
        pass
    try:
        YH = switch.info['fiftyTwoWeekHigh'] # 52 Week High Price
    except:
        pass
    try:
        CP = switch.info['currentPrice'] # Current Price
    except:
        pass
    try:       
        if DE < 40:
        # if PEG > 0 and PM > 0 and DE < 40:
            print(item, MC, AV, DE, BETA, PM, ROA, PR, DY, CR, RM, TPE, FPE, PEG, PB, TMP, YL, YH, CP)        
    except:
        pass
    
print()

# End calculating runtime
end_time = time.monotonic()
print("Runtime: ", timedelta(seconds=end_time - start_time))
```
